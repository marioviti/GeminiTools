```python
import sys
from pathlib import Path

ROOT = Path.cwd().parent          # VisionModelsRepo/
SRC = ROOT / "src"

sys.path.insert(0, str(SRC))

import gemini_tools
```

```python
from PIL import Image

image = Image.open("./image_example.webp")
```

```python
from gemini_tools.model import GeminiLLM
from google.genai import types

genai_key = input("Enter your genai key: ")

model = GeminiLLM(api_key=genai_key)

response_schema = types.Schema(
    type=types.Type.OBJECT,
    properties={
        "objects": types.Schema(
            type=types.Type.ARRAY,
            items=types.Schema(
                type=types.Type.OBJECT,
                properties={
                    "noun": types.Schema(type=types.Type.STRING),
                    "attributes": types.Schema(
                        type=types.Type.OBJECT,
                        properties={
                            "color": types.Schema(type=types.Type.STRING),
                            "depth": types.Schema(
                                type=types.Type.STRING,
                                enum=["foreground", "background", "middle_ground"],
                            ),
                            "vertical": types.Schema(
                                type=types.Type.STRING,
                                enum=["top", "bottom", "center"],
                            ),
                            "horizontal": types.Schema(
                                type=types.Type.STRING,
                                enum=["left", "right", "center"],
                            ),
                            "container" : types.Schema(type=types.Type.BOOLEAN),
                            "shape": types.Schema(type=types.Type.STRING),
                        },
                        required=[
                            "color",
                            "container",
                            "depth",
                            "vertical",
                            "horizontal",
                            "shape",
                        ],
                    ),
                },
                required=["noun", "attributes"],
            ),
        )
    },
    required=["objects"],
)

model.generate(["describe food, return max 5 objects", image], max_output_tokens=1024, response_schema=response_schema)
```
